```python
import pandas as pd
```

```python
data = pd.read_csv("/content/exams.csv")
```

```python
data['math score'].fillna(data['math score'].mean(), inplace=True)
```

```python
data['reading score'].fillna(data['reading score'].mean(), inplace=True)
```

```python
data['writing score'].fillna(data['writing score'].mean(), inplace=True)
```

```python
data['gender'].fillna(data['gender'].mode()[0], inplace=True)
```

```python
data['race/ethnicity'].fillna(data['race/ethnicity'].mode()[0], inplace=True)
```

```python
data['parental level of education'].fillna(data['parental level of education'].mode()[0], inplace=True)
```

```python
data['lunch'].fillna(data['lunch'].mode()[0], inplace=True)
```

```python
data['test preparation course'].fillna(data['test preparation course'].mode()[0], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
```

```python
data['gender'] = le.fit_transform(data['gender'])
data['race/ethnicity'] = le.fit_transform(data['race/ethnicity'])
data['parental level of education'] = le.fit_transform(data['parental level of education'])
data['lunch'] = le.fit_transform(data['lunch'])
data['test preparation course'] = le.fit_transform(data['test preparation course'])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import  train_test_split
from sklearn.metrics import accuracy_score

# Split features and target
X = data.drop(columns=['writing score'])
y = data['writing score']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Define the Decision Tree classifier with desired parameters
classifier = DecisionTreeClassifier(max_depth=5, criterion='gini')

# Train the classifier
classifier.fit(X_train, y_train)

# Predict on the test set
y_pred = classifier.predict(X_test)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = model.predict(X_test)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print("Model Evaluation:")
print("RMSE:", rmse)
print("R2 Score:", r2)
```

These results suggest that the Linear Regression model is performing quite well, with a low RMSE and a high R2 score. It means that the model is able to explain a significant portion (93.54%) of the variance in the Writing Score based on the features provided.
